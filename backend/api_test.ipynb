```python
import requests
import mysql.connector
import time

# 配置 MySQL 连接
db_config = {
    'user': 'root',
    'password': 'root',
    'host': 'localhost',
    'database': 'summer_project'
}

# 获取 jzgbh 列表
def get_jzgbh_list():
    connection = mysql.connector.connect(**db_config)
    cursor = connection.cursor()
    cursor.execute("SELECT DISTINCT JZGBH FROM papers_teachers_relationship")
    jzgbh_list = [row[0] for row in cursor.fetchall()]
    cursor.close()
    connection.close()
    return jzgbh_list
jzgbh_list = get_jzgbh_list()
# 取jzgbh列表的前100个
jzgbh_list = jzgbh_list[:20]
```

```python
jzgbh_list=[1, 7, 8, 9, 10, 15, 16, 19, 20, 23]
def test_batch_requests(jzgbh_list):
    url = "http://localhost:5000/api/Teacher_Summary"
    times = []

    for jzgbh in jzgbh_list:
        params = {'JZGBH': jzgbh}
        start_time = time.time()
        response = requests.get(url, params=params)
        end_time = time.time()
        elapsed_time = end_time - start_time
        times.append(elapsed_time)
        print(f"JZGBH: {jzgbh}, Status Code: {response.status_code}, Time: {elapsed_time:.4f} seconds")


if __name__ == "__main__":
    test_batch_requests(jzgbh_list)
```

```python
import redis
import json

# 配置 Redis 连接
try:
    redis_client = redis.Redis(host="localhost", port=6379, db=0)
except redis.ConnectionError as e:
    print(f"Redis connection error: {e}")
    exit(1)

# 检查特定教师的缓存
jzgbh = "23"
cache_key = f"teacher_dandelion:{jzgbh}"

if redis_client.exists(cache_key):
    summary_json = redis_client.get(cache_key)
    summary = json.loads(summary_json)
    print("缓存存在，内容如下：")
    print(json.dumps(summary, indent=4, ensure_ascii=False))
else:
    print("缓存不存在")
```
